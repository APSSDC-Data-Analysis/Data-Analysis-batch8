### 1. Line graph
- plt.plot()

```python
x = [1,2,3]
y = [10,11,12]
# (1,10), (2,11), (3,12)
plt.plot(x,y)
```

```python
x_axis = [10,12,14,16]
y_axis = [5,6,7,8]
plt.plot(x_axis, y_axis, label = 'line')
plt.title('Line Graph')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.legend()
```

```python
help(plt.legend)
```

```python
plt.plot(x_axis, y_axis, label = 'line')
plt.legend(loc = 'lower right')
```

```python
plt.plot(x_axis, y_axis, label = 'line')
plt.legend(loc = 10)
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,20)
x
```

```python
plt.plot(x, x**2, color = 'red', label = 'Square')
plt.plot(x, x**3, color = 'purple', label = 'Cube')
plt.plot(x, x**4, color = 'green', label = 'power4')
plt.legend()
```

```python
help(plt.plot)
```

### Subplots

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (12,4))
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (12,4))
ax1.plot(x, x)
ax2.plot(x, x**2)
ax3.plot(x, x**3)
```

### 2. Bar Graph
- plt.bar()

```python
Year = ['2019 Dec', '2020 Feb', '2020 June','2020 Oct']
Goldprice = [1000, 3000, 6000, 4000]
plt.bar(Year, Goldprice, color = 'g')
```

```python
# x = [0,1,2,3,4]
# y = [80,50,66,90,77]
# y1 = [20,30,40,50,60]
plt.bar([0,1,2,3,4],[80,50,66,90,77],width = 0.33)
plt.bar([0,1,2,3,4], [20,30,40,50,60], width = 0.33)
```

```python
plt.bar([0,1,2,3,4], [80,90,40,50,66], width = 0.3)
plt.bar([0,1,2,3,4]+0.3, [22,33,92,44,55], width = 0.3)
```

```python
Year = ['2019 Dec', '2020 Feb', '2020 June','2020 Oct']
Goldprice = [1000, 3000, 6000, 4000]
plt.bar(Year, Goldprice, color = ['r','gray','green','m'], edgecolor = 'blue')
```
